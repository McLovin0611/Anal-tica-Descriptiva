# Análisis descriptivo

A continuación, vamos a estar haciendo un análisis descriptivo a un conjunto de datos que trata de los préstamos del Banco Agrícola de la República Dominicana (BAGRICOLA), abarcando el periodo 2017-2022.

```python
# Primeramente, vamos a exporta las librerías que necesitamos.

from pandas import *
from numpy import *
from matplotlib.pyplot import *
from seaborn import *
```

```python
# Ahora exportaremos el conjunto de datos (dataset) haciendo uso de la librería Pandas.

dataset = read_excel("Cartera de Préstamos, 2017 - 2022.xlsx", usecols = "A:E")
```

![Imagen](https://i1.wp.com/agricultura.gob.do/wp-content/uploads/2020/09/LogoBagricola.png?w=225&ssl=1)

Este conjunto contiene datos sobre la Cartera de Préstamos del Banco Agrícola de la República Dominicana (BAGRICOLA), abarcando el periodo 2017-2022, en el cual se puede encontrar información general sobre los préstamos otorgados por la institución.

__Nota:__ Una cartera de préstamos es un conjunto de créditos y financiamientos que los bancos, compañías de inversión poseen o manejan.

```python
# Llamamos a nuestro conjunto de datos (dataset).

dataset
```

```python
filas = dataset[["SUCURSAL","CANTIDAD DE PRESTAMOS", "AÑO"]]

filas
```

```python
cantidad_de_prestamos = dataset[dataset["AÑO"] == 2022]

cantidad_de_prestamos
```

```python
# A continuación, vamos a ver la cantidad total de prestamos que se realizaron en el año actual 2022.

cantidad_de_prestamos["CANTIDAD DE PRESTAMOS"].sum()
```

```python
# Vamos a realizar un gráfico de barra horizontal que nos muestre el total de prestamos por sucursal en el año actual (2022).

grafico1 = figure(figsize = (16, 8))
title("Cantidad de préstamos realizados por sucursales en el año actual (2022)")

(cantidad_de_prestamos.groupby(["SUCURSAL"]) ["CANTIDAD DE PRESTAMOS"].sum()).plot(kind = "barh")
xlabel("CANTIDAD DE PRESTAMOS")
ylabel("SUCURSALES")
show()
```
